```python
# CANCELLATION RATE ANALYSIS
# Count each ride status
status_count = uber['ride_status'].value_counts()
```

```python
plt.figure(figsize=(9, 5))
colors = ['green' if i == 'Completed' else 'red' for i in status_count.index]
plt.bar(status_count.index,status_count.values,color=colors,edgecolor='black',label='Ride Status')
plt.title("Ride Status Distribution", fontsize=16, fontweight='bold')
plt.xlabel("Ride Status", fontsize=13, fontweight='bold')
plt.ylabel("Number of Rides", fontsize=13, fontweight='bold')
plt.legend(['Completed = Green | Cancelled = Red'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
total_rides      = len(uber)
cancelled_rides  = len(uber[uber['ride_status'] == 'Cancelled'])
completed_rides  = len(uber[uber['ride_status'] == 'Completed'])
cancel_rate      = (cancelled_rides / total_rides) * 100
complete_rate    = (completed_rides / total_rides) * 100 
print(f"Total Rides: {total_rides}")
print(f"Completed Rides: {completed_rides} ({complete_rate:.2f}%)")
print(f"Cancelled Rides: {cancelled_rides} ({cancel_rate:.2f}%)")
```
